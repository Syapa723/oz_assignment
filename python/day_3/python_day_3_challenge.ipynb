```

```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]

nums_min = nums [0]
nums_max = nums [8]
nums_avg = sum(nums) / len(nums)

print(f'리스트의 최댓값은 {nums_max}, 최솟값은 {nums_min}, 그리고 평균값은 {nums_avg}입니다.')
````
