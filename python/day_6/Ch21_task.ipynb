사용자로부터 이름과 나이를 입력 받아, "안녕하세요, [이름]님! 당신은 [나이]살입니다."라고 출력하는 함수 greet_user를 작성하세요.

```python
class person :
    def __init__(self, name, age) :
        self.hello = '안녕하세요,'
        self.name = name
        self.age = age

    def greet_user(self) :
        print('{0} {1}님! 당신은{2}살 입니다.'.format(self.hello, self.name, self.age))

user = person('오즈','26')

user.greet_user()
```
